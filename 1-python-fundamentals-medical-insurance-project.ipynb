## 4. Python Loops

###  To analyze medical insurance cost data efficiently without writing repetitive code, we will use Python loops to iterate through and analyze medical insurance cost data.

Our data is stored in the following three lists:

```python
names = ["Judith", "Abel", "Tyson", "Martha", "Beverley", "David", "Anabel"]
estimated_insurance_costs = [1000.0, 2000.0, 3000.0, 4000.0, 5000.0, 6000.0, 7000.0]
actual_insurance_costs = [1100.0, 2200.0, 3300.0, 4400.0, 5500.0, 6600.0, 7700.0]
```

 let's calculate the average insurance cost each person paid.

```python
# We’ll start by adding up all of the insurance costs.
# let's first create a variable total_cost and initialize it to 0, then use a for loop to iterate through actual_insurance_costs
# and add each insurance cost to the variable total_cost

total_cost = 0
for i in actual_insurance_costs:
  total_cost += i
```

```python
# Let's  create a variable called average_cost that stores the total_cost divided by the length of the 
# actual_insurance_costs list and print it.

average_cost = total_cost/(len(actual_insurance_costs))

print('Average Insurance Cost: ' + str(average_cost) + ' dollars.')
```

```python
# For each individual in names, we want to determine whether their insurance cost is above or below average. Let's write 
# a for loop with variable i that goes from 0 to len(names) and do the following inside the for loop:
# Create a variable name, which stores names[i].
# Create a variable insurance_cost, which stores actual_insurance_costs[i].
# use if, elif, else statements to check whether the insurance cost is above, below, or equal to the average. 
# and print out messages for each case.

for i in range(len(names)):
  name = names[i]
  insurance_cost = actual_insurance_costs[i]
  if insurance_cost > average_cost:
    print('The insurance cost for ' + name + ' is above average.')
  elif insurance_cost < average_cost:
    print('The insurance cost for ' + name + ' is below average.')
  else:
    print('The insurance cost for ' + name + ' is equal to the average.')

```

```python
# If we look closely at actual_insurance_costs and estimated_insurance_costs, we notice that each of the actual insurance
# costs are 10% higher than the estimated insurance costs.
# Using a list comprehension, let's create a new list called updated_estimated_costs, which has each element in 
# estimated_insurance_costs multiplied by 11/10 and print the new list.

updated_estimated_costs = [i * 11/10 for i in estimated_insurance_costs]
print(updated_estimated_costs)
```

## 5. Python Strings

### As a doctor who needs to clean up medical patient records, which are currently stored in one large string, we will use  knowledge of Python strings to obtain and clean up medical data so that it is easier to read and analyze.

The following string named medical_data stores the medical records for ten individuals. Each record is separated by a ; and contains the name, age, BMI (body mass index), and insurance cost for an individual, in that order.

```python
medical_data = \
"""Marina Allison   ,27   ,   31.1 , 
#7010.0   ;Markus Valdez   ,   30, 
22.4,   #4050.0 ;Connie Ballard ,43 
,   25.3 , #12060.0 ;Darnell Weber   
,   35   , 20.6   , #7500.0;
Sylvie Charles   ,22, 22.1 
,#3022.0   ;   Vinay Padilla,24,   
26.9 ,#4620.0 ;Meredith Santiago, 51   , 
29.3 ,#16330.0;   Andre Mccarty, 
19,22.7 , #2900.0 ; 
Lorena Hodson ,65, 33.1 , #19370.0; 
Isaac Vu ,34, 24.8,   #7045.0"""

```

```python
# We want the insurance costs to be represented in US dollars. Let's replace all instances of # in 
# medical_data with $ and store the result in a variable called updated_medical_data.

updated_medical_data = medical_data.replace('#','$')

print(updated_medical_data)
```

```python
# We want to calculate the number of medical records in our data.
# Let's create a variable called num_records and initialize it at 0 and then write a for loop to iterate through the 
# updated_medical_data string. Inside of the loop, add 1 to num_records when the current character is equal to $.

num_records=0
for i in updated_medical_data:
  if i == '$':
    num_records +=1
print('There are ' + str(num_records) + ' medical records in the data.')
```

```python
# The medical data in its current form is difficult to analyze. An essential job for a data scientist is to clean up data 
# so that it’s easy to work with.

# Let’s start off by splitting the updated_medical_data string into a list of each medical record. Remember that each
# medical record is separated by a ; in the string. Let's store the result in a variable called medical_data_split and 
# print this variable.

medical_data_split=updated_medical_data.split(';')

print(medical_data_split)
```

```python
# Our data is now stored in a list, but it is still hard to read. Let’s split each medical record into its own list.
# For that, let's first, define an empty list called medical_records and then, iterate through medical_data_split 
# and for each record, split the string after each comma (,) and append the split string to medical_records.

medical_records = []
for i in medical_data_split:
  medical_records.append(i .split(','))
print(medical_records)
```

```python
# Our data is now slightly more readable. However, it is not properly formatted – it contains unnecessary whitespace.
# To fix this, let’s start by creating an empty list called medical_records_clean.
# Next, use a for loop to iterate through medical_records.
#Inside of the loop, create an empty list called record_clean. We’ll use this list to store a formatted version of each
# medical record.
# After the record_clean variable, create a nested for loop that goes through each record.
# Inside of this loop, append item.strip() to record_clean to remove any whitespace from the string.

medical_records_clean=[]
for record in medical_records:
  record_clean = []
  for item in record:
    record_clean.append(item.strip())

# Finally, we need to add each cleaned up record to medical_records_clean.
# Outside of the nested for loop, let's append record_clean to medical_records_clean.

  medical_records_clean.append(record_clean)
print(medical_records_clean)
```

```python
# Our data is now clean and ready for analysis. Let's print out the names of each of the ten individuals, with all characters 
# uppercase, by using for loop.

for record in medical_records_clean:
  record[0] = record[0].upper()
  print(record[0])
```

```python
# Let’s store each name, age, BMI, and insurance cost in separate lists.To start, create four empty lists:
# names, ages, bmis, insurance_costs.
# Next, we will iterate through medical_records_clean and for each record we will:
# Append the name to names, age to ages, BMI to bmis and the insurance cost to insurance_costs.


names = []
ages = []
bmis = []
insurance_costs = []

for record in medical_records_clean:
  # append the name, age, BMI, and insurance cost from the current medical record 
  names.append(record[0])
  ages.append(record[1])
  bmis.append(record[2])
  insurance_costs.append(record[3])
```

```python
# Finally, let's print our output


print("Names: " + str(names))
print("Ages: " + str(ages))
print("BMI: "  + str(bmis))
print("Insurance Costs: " + str(insurance_costs))
```

```python
# Now that all of our data is in separate lists, we can easily perform analysis on that data. 
# Let’s calculate the average BMI in our dataset.

total_bmi = 0

for bmi in bmis:
  total_bmi += float(bmi)
average_bmi = total_bmi/len(bmis)
print("Average BMI: " + str(average_bmi))
```
