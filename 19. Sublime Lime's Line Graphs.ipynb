### Sales Data Analyses using Line Graphs.

In this project, we will be acting as a data analyst for an online lime retailer called Sublime Limes. People all over the world can use this product to get the freshest, best-quality limes delivered to their door. One of the product managers at Sublime Limes would like to gain insight into the customers and their sales habits. Using Matplotlib, weâ€™ll create some different line graphs (or maybe, lime graphs) to communicate this information effectively.

```python
# We will first import the library for plotting graphs.

from matplotlib import pyplot as plt
```

Let's look through the following lists and try to understand what each one represents.

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]
```

numbers of limes of different species sold each month:

```python
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]

persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]

blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

#### Plotting the Line Graphs:

```python

# We are going to make two charts in one figure, laid out side-by-side.

plt.figure(figsize=(12,8)) # Create a figure of width 12 and height 8. 

ax1=plt.subplot(1,2,1) # command to create the left subplot. We are going to plot the total page visits over the past year as a line.
x_values=range(len(months)) # creating the list of x-values, which is range(len(months)).
plt.plot(x_values, visits_per_month, marker='o') # plotting the total page visits against these x_values as a line and give the line markers that will help show each month as a distinct value.
plt.xlabel('Months') # labeling the x-axis and y-axis with descriptive titles of what they measure.
plt.ylabel('Total page visits')
ax1.set_xticks(x_values) # setting the x-axis ticks to be the x_values.
ax1.set_xticklabels(months) # labeling the x-axis tick labels to be the names stored in the months list.
plt.title('Total Visitors per Month') # Giving a descriptive title to the plot.


# In the subplot on the right, we are going to plot three lines on the same set of axes. The x-values for all three lines will
# correspond to the months, so we can use the list of x_values we used for the last plot.
# On this plot, let's create the three lines:

# 1. number of key limes sold vs x_values
# 2. number of Persian limes sold vs x_values
# 3. number of blood limes sold vs x_values

ax2=plt.subplot(1,2,2)
plt.plot(x_values, key_limes_per_month, color='yellow', marker='o') # Let's also give each line a specific color.
plt.plot(x_values, persian_limes_per_month, color='purple', marker='s')
plt.plot(x_values, blood_limes_per_month, color='red', marker='*')
plt.legend(['key limes', 'persian limes', 'blood limes']) # Add a legend to differentiate the lines, labeling each lime species.
ax2.set_xticks(x_values) # Set the x-axis ticks to be the x_values.
ax2.set_xticklabels(months) # Set the tick labels to be the months list.
plt.title('Different lime species sold per month') # Adding a descriptive title.
plt.savefig('lime_graphs.png') # Saving our figure as a png with a descriptive file name.
plt.show()
```

#### Conclusion:

1. From the first graph between Total page visits in different months, we observe that the maximum number of page visits occur in the months of June and July, which is pehaps due to the hot weather. 

2. From the graph on the  right, which shows the number of different lime spesies sold per month, we observe that key limes are the most popular throughout the year with maximum demand in the months from Jan to April, Persian limes have the least demand in most of the months, except in the months of May and June during which they have a demand comparable to key limes. Blood limes on the other hand have almost steady demand, which is inbetween that of the other two types of lims throughout the year, which decreases in the month of september and then rises again till December.
