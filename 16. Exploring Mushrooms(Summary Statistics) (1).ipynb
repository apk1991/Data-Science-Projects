## Exploring Mushrooms

Mushrooms exist in a variety of different colors, shapes, sizes, textures, etc. In this project, we will analyze an extensive mushroom dataset from UCI(https://www.kaggle.com/uciml/mushroom-classification) using bar charts and acquaint ourselves with the diverse array of mushrooms that exist worldwide.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Let's explore the dataset.

df = pd.read_csv(r"C:\Users\amanp\OneDrive\Desktop\mushroom_data.csv")
df.head()
```

We can see that the dataset contains 23 columns of data describing thousands of mushrooms. If we read through the header of dataset, we see that every single variable is a categorical variable. Therefore, we can create bar charts for the counts of every variable in the csv file.


There are 23 variables in this dataset (one for each column). Graphing each one individually would be tedious; so, will use loops.


```python
# Let's first create an attribute called columns. This attribute returns the name of each variable in mushroom_data.csv
columns = df.columns.tolist()
```

```python
# Let's create a loop that traverses each column in the columns list.
# then plot our data using the .countplot() method from the seaborn library.
# We will call .countplot() in the for loop.
# Let's use plt.clf() method after plt.show().This will keep any graphs we are plotting from bunching up on each other. 
# Instead, our plots will be neatly stacked on top of each other will spacing ideal for viewing.

for column in columns:
    sns.countplot(df[column])
    plt.show()
    plt.clf()
```

Good job! We have got 23 graphs. Let's clean up the graphs and make them more readable and useful for finding patterns.

```python
# 1. First, if we look at the graphs, we see that some of the x-axis labels overlap each other.
# 2. Also, the font size for the labels along the x-axis is also pretty small, making them tough to read.
# let's fix these two issues and make the graphs again.


for column in columns:
    sns.countplot(df[column])
    # rotates the value labels slightly so they donâ€™t overlap, also slightly increases font size
    plt.xticks(rotation=30, fontsize=10)
    # increases the variable label font size slightly to increase readability
    plt.xlabel(column, fontsize=12)
    plt.show()
    plt.clf()
```

```python
# 3. Next, to increase readability, let's add informative title.

for column in columns:
    sns.countplot(df[column])
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(column, fontsize=12)
    # we will use column to capture each variable name.
    plt.title(column + " Value Counts")
    plt.show()
    plt.clf()
```

```python
# 4. The graphs are readable, but let's take another step further.
# Let's add the order parameter to our .countplot() method and set this parameter so that the value counts in each column are
# in descending order.

for column in columns:
    # we will use the .value_counts() pandas method and the .index pandas object
    sns.countplot(df[column], order=df[column].value_counts().index)
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(column, fontsize=12)
    plt.title(column + " Value Counts")
    plt.show()
    plt.clf()

```
